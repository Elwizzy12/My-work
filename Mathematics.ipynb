```python
# Basic Quadratic Function: Implement and plot $g(x) = x^2 + 2x + 5$.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate x-values using np.linspace
x_values = np.linspace(-10, 10, num=100)
```

```python
y_values = x_values**2 + 2*x_values + 5
```

```python

# Plot the function
plt.plot(x_values, y_values)

# Set a title for the plot
plt.title("Plot of g(x) = x^2 + 2x + 5")

# Set labels for the x-axis and y-axis
plt.xlabel("x")
plt.ylabel("g(x)")

# Turn on the grid lines
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Sine Function: Implement the basic sine function $h(x) = \sin(x)$ and visualize it.
```

```python
x_values = np.linspace(-10, 10, num=100)
```

```python

# Generate x-values using np.linspace
x_values = np.linspace(-2*np.pi, 2*np.pi, 100)
```

```python
# Calculate y-values for the function h(x)
y_values = np.sin(x_values)

# Plot the function
plt.plot(x_values, y_values)

# Set a title for the plot
plt.title("Plot of h(x) = sin(x)")

# Set labels for the x-axis and y-axis
plt.xlabel("x")
plt.ylabel("h(x)")

# Turn on the grid lines
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Combine Quadratic & Sine: Implement $i(x) = \sin(x^2 + 2x + 5)$ and visualize.
```

```python
x_values = np.linspace(-10, 10, num=100)
```

```python
y_values = np.sin(x_values**2 + 2*x_values + 5)
```

```python
plt.plot(x_values, y_values)

# Set a title for the plot
plt.title("Plot of i(x) = sin(x^2 + 2x + 5)")

# Set labels for the x-axis and y-axis
plt.xlabel("x")
plt.ylabel("i(x)")

# Turn on the grid lines
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Exponential Decay: Visualize the function $j(x) = e^{-0.1x}$ using the exponential function.
```

```python
x_values = np.linspace(-10, 10, num=100)
```

```python
y_values = np.exp(-0.1*x_values)
```

```python
# Plot the function
plt.plot(x_values, y_values)

# Set a title for the plot
plt.title("Plot of j(x) = e^(-0.1x)")

# Set labels for the x-axis and y-axis
plt.xlabel("x")
plt.ylabel("j(x)")

# Turn on the grid lines
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Complete Function: Combine all the above components to implement and visualize $f(x) = \sin(x^2 + 2x + 5)$ times $e^{-0.1x}$.
```

```python
x_values = np.linspace(-10, 10, num=100)
```

```python
# Calculate y-values for the function f(x)
y_values = np.sin(x_values**2 + 2*x_values + 5) * np.exp(-0.1*x_values)
```

```python
plt.plot(x_values, y_values)

# Set a title for the plot
plt.title("Plot of f(x) = sin(x^2 + 2x + 5) * e^(-0.1x)")

# Set labels for the x-axis and y-axis
plt.xlabel("x")
plt.ylabel("f(x)")

# Turn on the grid lines
plt.grid(True)

# Display the plot
plt.show()
```
